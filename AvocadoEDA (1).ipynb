```python
import pandas as pd
```

```python
avocados = pd.read_csv('./avocados.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
avocados["Date"] = pd.to_datetime(avocados["Year "].astype(str) + " " + avocados["Month"])
```

```python
plt.figure(figsize=(10,6))
plt.plot(avocados["Date"], avocados["ABP (Adjusted Base Price)"], marker="o", label="Adj. Based Price")
plt.plot(avocados["Date"], avocados["PP (Parity Price)"], marker="s", label="Parity Price")

plt.title("ABP and PP Over Time")
plt.xlabel("Date")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,8), sharex=True)

ax1.plot(avocados["Date"], avocados["ABP (Adjusted Base Price)"], 
         marker="o", color="tab:blue", label="ABP")
ax1.set_ylabel("ABP (Adjusted Base Price)", color="tab:blue")
ax1.set_title("ABP Over Time")
ax1.grid(True)
ax1.legend()

ax2.plot(avocados["Date"], avocados["PP (Parity Price)"], 
         marker="s", color="tab:red", label="PP")
ax2.set_ylabel("PP (Parity Price)", color="tab:red")
ax2.set_title("PP Over Time")
ax2.set_xlabel("Date")
ax2.grid(True)
ax2.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
